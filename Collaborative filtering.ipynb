```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies=pd.read_csv('dataset/movies.csv')
ratings=pd.read_csv('dataset/ratings.csv')
```

```python
ratings = pd.merge(ratings,movies,on='movieId')
```

```python
ratings.head()
```

```python
movies.head()
```

```python
avg_ratings_movie = pd.DataFrame(ratings.groupby('title')['rating'].mean())
```

```python
avg_ratings_movie.head()
```

```python
avg_ratings_movie['no_of_ratings'] = pd.DataFrame(ratings.groupby('title')['rating'].count())
```

```python
avg_ratings_movie.head()
```

```python
rating_pivot_table = ratings.pivot_table(index='userId',columns='title',values='rating')
```

```python
rating_pivot_table.head()
```

```python
avg_ratings_movie.sort_values('no_of_ratings',ascending=False).head(10)
```

```python
def get_name():
    name = input("enter movie name : ")
    list=[]
    for item in movies['title']:
        if name in item:
            list.append(item)
    if len(list)==0:
        print("NO match found")
        get_name()
    elif len(list)==1:
        return list[0]
    else:
        print("Multiple Search Result")
        [print(x) for x in list]
        actual_name = input("Enter Exact name")
        return actual_name
```

```python
movie_ratings=rating_pivot_table[get_name()]
movie_ratings.head()
```

```python
similar_movie_ratings = rating_pivot_table.corrwith(movie_ratings)
```

```python
corr_movie = pd.DataFrame(similar_movie_ratings,columns=['correlation'])
corr_movie.dropna(inplace=True)
corr_movie.sort_values('correlation',ascending=False).head()
```

```python
corr_movie = corr_movie.join(avg_ratings_movie['no_of_ratings'])
```

```python
corr_movie.head()
```

```python
result = corr_movie[corr_movie['no_of_ratings']>50].sort_values('correlation',ascending=False).head(6)
```
